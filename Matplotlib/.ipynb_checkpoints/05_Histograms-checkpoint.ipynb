# üìä Histograms in Matplotlib

A **histogram** is a type of plot that shows the **distribution** of a dataset. It‚Äôs especially useful for visualizing the **frequency of numerical data** within specified ranges (called **bins**).

---

## ‚ùì Why and When to Use Histograms?

Use histograms when:

* üìà You want to understand the **distribution of a numerical dataset** (e.g. age, salary, views).
* üîç You want to detect **skewness**, **outliers**, or understand **spread**.
* üìè You‚Äôre **binning continuous data** into intervals.

### ‚úÖ Examples include:

* üë• Analyzing the **age of your YouTube viewers**
* üìö Understanding **test scores distribution**
* üß™ Checking if data is **normally distributed**

---

## üßÆ Understanding the `bins` Argument

The `bins` argument controls how the data is grouped:

* If an **integer**, it defines the number of **equal-width bins**.
* If a **list**, it defines **custom bin edges**, allowing you to control the range and width of each bin.

```python
plt.hist(data, bins=10)  # üîüEqual-width bins
plt.hist(data, bins=[10, 20, 30, 40, 60, 100])  # üéØ Custom age bins
```

---

## üé® `edgecolor` for Better Visibility

The `edgecolor` parameter adds **borders to the bars**, improving clarity:

```python
plt.hist(data, bins=10, edgecolor='black')
```

---

## üìò Example: Age Distribution of YouTube Viewers

Here is age data for some viewers:

```python
import matplotlib.pyplot as plt
import numpy as np
 
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

---

## üìç Adding a Vertical Line: `axvline`

Use `axvline` to **mark a specific value** like the average age or a threshold:

```python
plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(np.mean(ages), color='red', linestyle='--', linewidth=2, label='Average Age')
plt.legend()
plt.title('Age Distribution with Mean Line')
plt.show()
```

---

## üìù Summary

| Parameter   | Purpose                        |
| ----------- | ------------------------------ |
| `bins`      | Number or custom edges of bins |
| `edgecolor` | Color around each bar          |
| `axvline`   | Vertical reference line        |

````python
import matplotlib.pyplot as plt
````

````python
data = [87, 76, 92, 65, 58, 79, 84, 90, 73, 68,
        81, 95, 62, 77, 69, 88, 71, 66, 83, 74,
        91, 64, 59, 80, 85, 97, 61, 70, 89, 67,
        93, 78, 75, 96, 45, 72, 60, 86, 63, 98,
        55, 94, 99, 57, 56, 53, 52, 54, 50, 100]

plt.hist(data, bins=5, edgecolor="black")       # 5 Equal-width bins
plt.xlabel("Marks")
plt.ylabel("Number of students")
````

````python
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=bins, edgecolor="black")
plt.title("Age Distribution of YouTube Viewrs")
plt.xlabel("Age Group")
plt.ylabel("Number of Viewers")
plt.grid(True, linestyle='--', alpha=0.5)
````

````python
plt.hist(ages, bins=bins, edgecolor="black")
plt.axvline(35, color="red", linestyle='--', linewidth=2, label='Average')
plt.title("Age Distribution of YouTube Viewrs")
plt.xlabel("Age Group")
plt.ylabel("Number of Viewers")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
````
