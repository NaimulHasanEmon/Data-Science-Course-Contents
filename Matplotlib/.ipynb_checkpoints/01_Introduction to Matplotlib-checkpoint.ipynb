# üìä Plotting Charts with Matplotlib

Today, we‚Äôll cover the basics of **Matplotlib** ‚Äî the most fundamental plotting library in Python. By the end of this tutorial, you'll know how to create clean and powerful plots step by step!

## What is `matplotlib.pyplot`? üé®

`matplotlib.pyplot` is a module in Matplotlib ‚Äî think of it as a **paintbrush** for your data. You‚Äôll often see it imported as `plt`:

```python
import matplotlib.pyplot as plt
```

This short alias (`plt`) saves you from typing out `matplotlib.pyplot` every time. Nice, right?

---

## What is `plt.show()`? üëÄ

`plt.show()` is used to **display** the plot. Without it, in scripts, you might not see the plot window! Here‚Äôs how it works:

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

---

## Interacting with the Plot üîç

Once a plot appears, you can interact with it! You can:

* **Zoom In/Out** üîÑ
* **Pan** around üß≠
* **Navigate history** with arrows ‚¨ÖÔ∏è‚û°Ô∏è
* **Reset to home** üè†

These features are **automatically included** in the plot window!

---

## Real Data Example: Sachin Tendulkar‚Äôs Runs Over Time üèè

Let‚Äôs create a plot showing Sachin Tendulkar's runs over the years!

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.show()
```

---

## Adding X and Y Labels üìù

Make your plot more readable by adding **labels** to the axes and a **title**:

```python
plt.plot(years, runs)
plt.xlabel("Year")  # Label for X-axis
plt.ylabel("Runs Scored")  # Label for Y-axis
plt.title("Sachin Tendulkar's Yearly Runs")  # Title of the plot
plt.show()
```

---

## Multiple Lines in One Plot üìà

Let‚Äôs compare **Virat Kohli** and **Virender Sehwag** to Sachin:

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()  # Show the legend
plt.show()
```

---

## Why **label** is Better than a List in `legend` üí°

### **Bad Practice:**

```python
plt.plot(years, kohli)
plt.plot(years, sehwag)
plt.legend(["Kohli", "Sehwag"])  # prone to mismatch
```

### **Better Practice:**

```python
plt.plot(years, kohli, label="Kohli")
plt.plot(years, sehwag, label="Sehwag")
plt.legend()  # Automatically picks up labels
```

---

## Using Format Strings ‚ú®

You can customize the **color**, **marker**, and **line style** of each plot!

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

---

## Color and Line Style Arguments üé®

You can set colors and line styles for each plot:

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

---

## Line Width and Layout Tweaks ‚öôÔ∏è

To improve readability, you can adjust **line width**, add a **grid**, and use **tight layout**:

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)  # Add grid
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

---

## Using Styles in Matplotlib üé®

Matplotlib offers some **built-in styles** to enhance the look of your plots. Here‚Äôs how to see and use them:

```python
print(plt.style.available)
```

Try some styles:

```python
plt.style.use("ggplot")  # Use ggplot style
# or
plt.style.use("seaborn-v0_8-bright")  # Use seaborn style
```

---

## XKCD Comic Style üòÜ

Matplotlib even lets you create plots that look like **XKCD comics**!

```python
with plt.xkcd():
    plt.plot(years, kohli, label="Kohli")
    plt.plot(years, sehwag, label="Sehwag")
    plt.title("Epic Battle of the Batsmen")
    plt.legend()
    plt.show()
```

---

## Visualizing Tons of Data ‚Äì What Crowded Looks Like üèÉ

Too much data can make your plot look confusing. Let‚Äôs see what happens when we plot **50 random lines**:

```python
import numpy as np

for i in range(50):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

## Final Tips for Beginners üåü

* Always **start with simple plots** to get the hang of it.
* **Add labels** and **legends** early ‚Äî they make plots much clearer.
* Use `plt.grid()` and `plt.tight_layout()` to improve readability.
* Try out **different styles** to find the one that fits your use case best.
