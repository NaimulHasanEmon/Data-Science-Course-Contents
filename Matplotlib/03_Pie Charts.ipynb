# üç∞ **Pie Charts in Matplotlib**

## ü•ß **Matplotlib Pie Charts**
Pie charts are a great way to visualize **part-to-whole relationships**. They're visually appealing but work best when there are **fewer categories**‚Äîtoo many slices can lead to **clutter**. üßë‚Äçüíª

---

## üé® **Setting a Style**
Matplotlib allows you to customize your plot‚Äôs look. You can easily set a style to make it more visually engaging!

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # Choose any style you like
````

---

## üìä **Basic Pie Chart Example**

Visualizing the career runs of Indian cricketers in a simple pie chart.

### **Data**:

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

---

## üåà **Custom Colors**

You can **customize the colors** of the pie slices by using either **color names** or **hex codes**.

```python
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

---

## üñãÔ∏è **Add Edges and Style Slices with `wedgeprops`**

Want to customize the **slice edges**? Use `wedgeprops` for more control!

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
plt.show()
```

> **Tip**: Google **‚Äúmatplotlib wedgeprops‚Äù** for additional customization options. üìö

[**Visit the official documentation**](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Wedge.html)

---

## üí• **Highlight a Slice Using `explode`**

Use **`explode`** to bring attention to one or more slices by pulling them out.

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

---

## üåü **Add Shadows for a 3D Feel**

Give your pie chart some **depth** with shadows!

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
plt.show()
```

---

## üîÑ **Start Angle**

You can control the **rotation** of the pie chart by setting a **starting angle**.

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=140
)
plt.show()
```

---

## üìâ **Show Percentages with `autopct`**

Want to display the **percentages** of each slice? Use the **`autopct`** parameter.

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Career Run Share")
plt.show()
```

## üö´ **Crowded Pie Chart ‚Äì Why to Avoid**

Be mindful when creating pie charts with **too many categories**. Here's an example that demonstrates why it's not ideal:

### Example of a **Crowded Pie Chart**:

```python
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```

### Why avoid it?

* **Hard to compare slice sizes visually**
* **Cluttered and confusing** üåÄ
* **No clear insight** ü§∑‚Äç‚ôÇÔ∏è

**Better alternatives**: **bar charts** or **horizontal bar charts** üìä.

---

## üìã **Summary of Useful Pie Chart Parameters**

| **Parameter**    | **Description**                                      |
| ---------------- | ---------------------------------------------------- |
| **`labels`**     | Labels for each slice                                |
| **`colors`**     | Customize slice colors (names or hex codes)          |
| **`explode`**    | Pull out slices for emphasis                         |
| **`shadow`**     | Adds **depth-like shadow** effect                    |
| **`startangle`** | Rotates pie to start from a **different angle**      |
| **`autopct`**    | Displays **percentage** text on slices               |
| **`wedgeprops`** | Customizes slice **edge**, **fill**, **width**, etc. |
