# üìä Basic Plot Types in Seaborn

Seaborn comes with **built-in example datasets**.
These are small real-world datasets like:

* üçΩÔ∏è *Restaurant tips*
* ‚úàÔ∏è *Flight passenger counts*
* üå∏ *Iris flower measurements*

They're great for:

* Practice
* Learning plotting techniques
* Avoiding manual data downloads

You can load these datasets directly into a **pandas DataFrame** using:

```python
sns.load_dataset()
```

---

## üîç How to See All Available Datasets

```python
import seaborn as sns

print(sns.get_dataset_names())
```

This will list dataset names like:
‚û°Ô∏è `tips`, `flights`, `iris`, `diamonds`, `penguins`, `titanic`, etc.

---

## üì• How to Load a Dataset

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

* This loads the **"tips"** dataset (restaurant bills and tips)
* Returns a **pandas DataFrame** ready for analysis or plotting

---

## ü§î Why Seaborn Provides Datasets

* ‚úÖ To quickly test different types of plots
* üìò To learn plotting without needing your own data
* üõ†Ô∏è To create examples and tutorials easily
* üßπ To demonstrate real-world data issues (e.g., **missing values**, **categorical data**)

---

## üìà Now Let‚Äôs Look Into Some Plots We Can Create Using Seaborn

---

### 1Ô∏è‚É£ Line Plot

```python
tips = sns.load_dataset('tips')

sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot Example')
plt.show()
```

---

### 2Ô∏è‚É£ Scatter Plot

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title('Scatter Plot with Color by Time')
plt.show()
```

---

### 3Ô∏è‚É£ Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Average Bill per Day')
plt.show()
```

---

### 4Ô∏è‚É£ Box Plot

Boxplots show **distributions**, **medians**, and **outliers** in one simple plot.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

---

### 5Ô∏è‚É£ Heatmap (Correlation Matrix)

```python
flights = sns.load_dataset('flights')
pivot_table = flights.pivot("month", "year", "passengers")

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```

## üß∞ Working with Pandas DataFrames

One of the biggest strengths of Seaborn:

```python
import pandas as pd

df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.show()
```
