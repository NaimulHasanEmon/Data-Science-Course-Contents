# üõ†Ô∏è Handling Missing Data with `SimpleImputer` from Scikit-Learn

We'll use **Option 3** using `SimpleImputer` from Scikit-Learn, which allows **consistent handling** across all datasets (`train`, `test`, `new data`) üöÄ:

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

housing_num = housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)
```

‚úÖ This computes the **median** for each numerical column and stores it in `imputer.statistics_`:

```python
>>> imputer.statistics_
array([-118.51 ,  34.26 ,  29. , 2125. , 434. , 1167. , 408. , 3.5385])
```

üìå Now apply the learned **medians** to transform the data:

```python
X = imputer.transform(housing_num)
```

---

## üìä Other Available Strategies:

* `"mean"` ‚Äì replaces with **mean value** üßÆ
* `"most_frequent"` ‚Äì replaces with the **most common value** (can handle **categorical**) üîÅ
* `"constant"` ‚Äì fill with a **fixed value** using `fill_value=` üß±

---

üîß Use the right strategy depending on the nature of your data for best results!

````python
import pandas as pd
import numpy as np
````

````python
data = pd.read_csv("xlsx/California Housing Prices/housing.csv")
````

````python
data["income_cat"] = pd.cut(data["median_income"], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])
````

````python
import matplotlib.pyplot as plt
````

````python
from sklearn.model_selection import StratifiedShuffleSplit
````

````python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
````

````python
for train_index, test_index in split.split(data, data["income_cat"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
````

````python
# Remove the income_cat column
for sett in (strat_train_set, strat_test_set):
    sett.drop("income_cat", axis=1, inplace=True)
````

````python
df = strat_train_set.copy()
````

````python
housing = strat_train_set.drop("median_house_value", axis=1)
housing_labels = strat_train_set["median_house_value"].copy()
````

````python
from sklearn.impute import SimpleImputer
````

````python
imputer = SimpleImputer(strategy="median")
````

````python
housing_num = housing.select_dtypes(include=[np.number])
````
